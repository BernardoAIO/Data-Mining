```python
import pandas as pd


df = pd.read_csv('classificacao_binaria_train.csv', sep=';', header=0)
```

```python
# Linhas duplicadas no DataFrame
duplicatas = df[df.duplicated()]
print("Linhas duplicadas:")
print(duplicatas)
```

Não há linhas duplicadas!

### Valores Ausentes

```python
valores_ausentes = df.isnull().sum()

# Colunas com valores ausentes e quantidade de valores ausentes em cada uma delas
colunas_com_valores_ausentes = valores_ausentes[valores_ausentes > 0]
print("Colunas com valores ausentes:")
print(colunas_com_valores_ausentes)

# Porcentagem de valores ausentes em cada coluna
porcentagem_valores_ausentes = (valores_ausentes / len(df)) * 100
print("\nPorcentagem de valores ausentes por coluna:")
print(porcentagem_valores_ausentes)
```

```python
import matplotlib.pyplot as plt


# Filtrar colunas com mais de 7% de valores ausentes
colunas_com_valores_ausentes = porcentagem_valores_ausentes[porcentagem_valores_ausentes > 7]

# Plotar o gráfico de barras
plt.figure(figsize=(12, 6))
colunas_com_valores_ausentes.plot(kind='bar', color='red')
plt.title('Valores Ausentes Consideráveis')
#plt.xlabel('Coluna')
plt.ylabel('Porcentagem')
plt.xticks(rotation=90)
plt.show()
```

Análise sobre valores ausentes nas colunas:  
*DT_NASCIMENTO* - Não deveria haver valores ausentes, porém, uma pequena quantidade dos registros dessa coluna são ausentes.  
*DS_TIPO_CONSULTA*  - Possui um grande número de registros ausentes e caracteriza se é primeira consulta, retorno ou encaminhamento do passageiro.  
*QT_TEMPO_DOENCA* - Grande número de valores ausentes e registra o tempo que o paciente está doente, além disso, é um campo numérico, ou seja, não deveria ter tantos valores ausentes.  
*DS_UNIDADE_TEMPO_DOENCA* - Acompanha a coluna anterior e registra a unidade de tempo que o paciente está doente (dias, meses, anos), também nao deveria apresentar muitos valores ausentes.  
*DS_TIPO_DOENCA* - Define se a doença é aguda ou crônica, entretanto, muitos valores ausentes.  
*DS_INDICACAO_ACIDENTE* - Indica qual tipo de acidente (trânsito, trabalho, etc). Apresenta um pequeno (comparado às outras colunas) número de valores ausentes.  
*DS_TIPO_SAIDA* - Todos valores ausentes e refere-se ao PROVÁVEL DESFECHO DO ATENDIMENTO: RETORNO, INTERNAÇÃO OU ALTA.  
*DS_TIPO_INTERNACAO* - Muitos valores ausentes e refere-se ao tipo de internação (CLÍNICA, CIRÚRGICA, OBSTÉTRICA, PSIQUIÁTRICA, ETC)  
*DS_REGIME_INTERNACAO* - Domiciliar ou hospitlar, muitos valores ausentes sem motivo aparente (mesmo número de dados faltantes da coluna anterior).  
*DS_TIPO_ACOMODACAO* - Apto, enfermaria ou UTI. Grande quantidade de valores faltantes sem motivo aparente.  
*QT_DIA_SOLICITADO* - Previsão de dias de internação, muitos dados faltantes, provavelmente por dificuldade de prever essa situação ou despreparo dos profissionais.  
*CD_GUIA_REFERENCIA* - Grande parte dos valores ausentes, porém, refere-se ao código da guia.  
*DS_TIPO_ATENDIMENTO* - Define se o atendimento é do tipo exame, pronto socorro, quimioterapia, etc. Parcela considerável de dados faltantes.  
*CD_CID* - Estabelece o CID que informa a provável doença que causou a solicitação, grande parte dos dados faltantes.  
*DS_INDICACAO_CLINICA* - Texto livre que define o principal motivo da solicitação. Parcela considerável de dados faltantes.  

As outras colunas não apresentam dados ausentes consideráveis, logo, não foram explicitadas.
