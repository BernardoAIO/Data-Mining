```python
# Update the unidimensional analysis function to handle columns with all null values
def unidimensional_analysis(column_name, data):
    analysis = {}

    # Check if the column contains only null values
    if data[column_name].isnull().all():
        analysis['type'] = 'All Nulls'
        analysis['message'] = 'The column contains only null values.'
        return analysis

    if data[column_name].dtype == 'object':
        # Categorical data
        analysis['type'] = 'Categorical'
        analysis['frequencies'] = data[column_name].value_counts()
        analysis['mode'] = data[column_name].mode()[0] if not data[column_name].mode().empty else 'No Mode'

        # Visualization: Bar plot
        plt.figure(figsize=(10, 6))
        sns.countplot(x=column_name, data=data, order=data[column_name].value_counts().index)
        plt.title(f'{column_name} - Frequencies')
        plt.xticks(rotation=45)
        plt.show()

    else:
        # Numerical data
        analysis['type'] = 'Numerical'

        # Measures of Location
        analysis['mean'] = data[column_name].mean()
        analysis['median'] = data[column_name].median()
        analysis['mode'] = data[column_name].mode()[0] if not data[column_name].mode().empty else 'No Mode'
        analysis['quantiles'] = data[column_name].quantile([0.25, 0.5, 0.75])

        # Measures of Dispersion
        analysis['std_dev'] = data[column_name].std()
        analysis['variance'] = data[column_name].var()
        analysis['range'] = data[column_name].max() - data[column_name].min()

        # Measures of Shape
        analysis['skewness'] = data[column_name].skew()
        analysis['kurtosis'] = data[column_name].kurt()

        # Visualization: Histogram and Boxplot
        plt.figure(figsize=(12, 6))

        plt.subplot(1, 2, 1)
        sns.histplot(data[column_name], bins=30, kde=True)
        plt.title(f'{column_name} - Distribution')

        plt.subplot(1, 2, 2)
        sns.boxplot(x=column_name, data=data)
        plt.title(f'{column_name} - Boxplot')

        plt.tight_layout()
        plt.show()

    return analysis
```

```python
# Function for bidimensional analysis
def bidimensional_analysis(column1, column2, data):
    analysis = {}

    if data[column1].dtype == 'object' and data[column2].dtype == 'object':
        # Categorical vs Categorical
        analysis['type'] = 'Categorical vs Categorical'

        # Cross-tabulation
        analysis['cross_tab'] = pd.crosstab(data[column1], data[column2])

        # Visualization: Heatmap of the cross-tabulation
        plt.figure(figsize=(12, 8))
        sns.heatmap(analysis['cross_tab'], annot=True, fmt='d', cmap='coolwarm')
        plt.title(f'{column1} vs {column2}')
        plt.show()

    elif (data[column1].dtype != 'object' and data[column2].dtype == 'object') or \
         (data[column1].dtype == 'object' and data[column2].dtype != 'object'):
        # Categorical vs Numerical or Numerical vs Categorical
        analysis['type'] = 'Categorical vs Numerical' if data[column1].dtype == 'object' else 'Numerical vs Categorical'

        # Visualization: Boxplot
        plt.figure(figsize=(12, 6))
        sns.boxplot(x=column1, y=column2, data=data) if data[column1].dtype == 'object' else sns.boxplot(x=column2, y=column1, data=data)
        plt.title(f'{column1} vs {column2}')
        plt.xticks(rotation=45)
        plt.show()

    else:
        # Numerical vs Numerical
        analysis['type'] = 'Numerical vs Numerical'

        # Correlation
        analysis['correlation'] = data[[column1, column2]].corr().iloc[0, 1]

        # Visualization: Scatter plot
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=column1, y=column2, data=data)
        plt.title(f'{column1} vs {column2} - Correlation: {analysis["correlation"]:.2f}')
        plt.show()

    return analysis

```

```python
# Update the bidimensional analysis function to include correlation coefficient for Numerical vs Numerical
def bidimensional_analysis(column1, column2, data):
    analysis = {}

    if data[column1].dtype == 'object' and data[column2].dtype == 'object':
        # Categorical vs Categorical
        analysis['type'] = 'Categorical vs Categorical'

        # Cross-tabulation
        analysis['cross_tab'] = pd.crosstab(data[column1], data[column2])

        # Visualization: Heatmap of the cross-tabulation
        plt.figure(figsize=(12, 8))
        sns.heatmap(analysis['cross_tab'], annot=True, fmt='d', cmap='coolwarm')
        plt.title(f'{column1} vs {column2}')
        plt.show()

    elif (data[column1].dtype != 'object' and data[column2].dtype == 'object') or \
         (data[column1].dtype == 'object' and data[column2].dtype != 'object'):
        # Categorical vs Numerical or Numerical vs Categorical
        analysis['type'] = 'Categorical vs Numerical' if data[column1].dtype == 'object' else 'Numerical vs Categorical'

        # Visualization: Boxplot
        plt.figure(figsize=(12, 6))
        sns.boxplot(x=column1, y=column2, data=data) if data[column1].dtype == 'object' else sns.boxplot(x=column2, y=column1, data=data)
        plt.title(f'{column1} vs {column2}')
        plt.xticks(rotation=45)
        plt.show()

    else:
        # Numerical vs Numerical
        analysis['type'] = 'Numerical vs Numerical'

        # Correlation using Pearson's coefficient
        analysis['correlation'] = data[[column1, column2]].corr(method='pearson').iloc[0, 1]

        # Visualization: Scatter plot
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x=column1, y=column2, data=data)
        plt.title(f'{column1} vs {column2} - Pearson Correlation: {analysis["correlation"]:.2f}')
        plt.show()

    return analysis

```
